### a) Åpne excel fil i Python
Last ned datasettet, pakk ut og legg filene under folderen data i ditt prosjekt.
Bruk deretter pandas sin read_excel funksjon til å lese filen Medals.xlsx.

Trinn-for-trinn instruksjoner:

- Les data fra Excel-filen

```python
# Skriv koden din her
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Steg 1: Last inn datasettet
file_path = '/Users/georgrommetveit/Documents/Inn1_anvendtprog/V-r2025/data/Medals.xlsx'
data = pd.read_excel(file_path)

print(data.head())
```

### b) Visualisere innholdet i Medals.xlsx

I denne oppgaven skal du visualisere innholdet i filen ved hjelp av Python-biblioteker som Matplotlib.
Bruk Matplotlib til å lage grafer som viser fordelingen av medaljer.    


```python
# Skriv koden din her
data_array = data.to_numpy()
countries = []
gold = np.zeros(len(data_array))
silver = np.zeros(len(data_array))
bronze = np.zeros(len(data_array))
total_medals = np.zeros(len(data_array))

print(data_array[0])
for i in range(len(data_array)):
    countries.append(data_array[i,1])
    gold[i] = data_array[i, 2]
    silver[i] = data_array[i, 3]
    bronze[i] = data_array[i, 4]
    total_medals[i] = data_array[i,5]



sorted_data = sorted(zip(total_medals, gold, silver, bronze, countries), reverse=True)
total_medals, gold, silver, bronze, countries = zip(*sorted_data)


plt.figure(figsize=(12, len(countries) * 0.6))  # Juster høyden basert på antall land

# Stablet stolpediagram
x_indices = np.arange(len(countries))
plt.barh(x_indices, gold, color='gold', label='Gold', height=0.5)
plt.barh(x_indices, silver, left=gold, color='silver', label='Silver', height=0.5)
plt.barh(x_indices, bronze, left=np.array(gold) + np.array(silver), color='#cd7f32', label='Bronze', height=0.5)

# Tilpasning av diagrammet
plt.yticks(x_indices, countries, fontsize=9)  # Landene på y-aksen
plt.xlabel('Number of Medals', fontsize=12)
plt.title('Olympic Medals by Country', fontsize=16)
plt.legend(title='Medal Type', fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)

for i, (g, s, b) in enumerate(zip(gold, silver, bronze)):
    total = g + s + b
    plt.text(total + 1, i, str(total), va='center', fontsize=9)

# Vis diagrammet
plt.tight_layout()
plt.show()

```

### c) Visualisere innholdet i EntriesGender.xlsx

I denne oppgaven skal du åpne filen EntriesGender.xlsx, som inneholder informasjon om kjønnsfordelingen av deltakere i forskjellige arrangementer under de olympiske leker i Tokyo 2021. Du skal deretter visualisere innholdet i filen ved hjelp av Python-biblioteker som Pandas og Matplotlib.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/Users/georgrommetveit/Documents/Inn1_anvendtprog/V-r2025/data/EntriesGender.xlsx'
gender_data = pd.read_excel(file_path)
gender_data_array = gender_data.to_numpy()

disciplines = []
female = np.zeros(len(gender_data_array))
male = np.zeros(len(gender_data_array))
total = np.zeros(len(gender_data_array))

for i in range(len(gender_data_array)):
    disciplines.append(gender_data_array[i, 0])
    female[i] = gender_data_array[i, 1]
    male[i] = gender_data_array[i, 2]
    total[i] = gender_data_array[i, 3]

x_indices2 = np.arange(len(disciplines))
plt.figure(figsize=(15, 8))  # Make the figure wider
plt.bar(x_indices2, female, label='Female')
plt.bar(x_indices2, male, bottom=female, label='Male')
plt.xticks(x_indices2, disciplines, fontsize=10, rotation=45, ha='right')  # Rotate labels for better spacing

plt.xlabel('Disciplines', fontsize=12)
plt.ylabel('Number of Participants', fontsize=12)
plt.title('Participants by Gender and Discipline', fontsize=14)
plt.legend()

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()



```
